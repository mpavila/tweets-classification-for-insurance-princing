```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt

# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["a","al","ante","así","bajo","como","con","contra","cuando","de","del","desde","el","él","ella","ellos","en","es","esa","ese","eso","esta","está","este","esto","estos","hay","la","las","le","les","lo","los","má","mi","mis","ni","nos","para","pero","por","porque","que","qué","RT","se","si","son","su","sus","te","tiene","todo","tu","un","una","unas","uno","unos","va","ya","yo",
])

text = " ".join(df['Tweet Text'])

# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords, max_font_size=70, max_words=200, background_color="white").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("datasets/wordcloud_corpus.png")
```
